```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de producción e inventarios
meses = np.arange(1, 7)
produccion = [205.0, 225.0, 190.0, 160.0, 225.0, 225.0]
inventario = [25.0, 0.0, 0.0, 20.0, 25.0, 0.0]

# Crear el gráfico de barras
fig, ax = plt.subplots()

# Gráfica de producción
ax.bar(meses - 0.2, produccion, width=0.4, label='Producción', color='blue')

# Gráfica de inventarios
ax.bar(meses + 0.2, inventario, width=0.4, label='Inventario', color='orange')

# Etiquetas y título
ax.set_xlabel('Meses')
ax.set_ylabel('Cantidad de ventanas')
ax.set_title('Producción e Inventario óptimos por mes')
ax.set_xticks(meses)
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Datos de demanda
demanda = [180, 250, 190, 140, 220, 250]

# Crear el gráfico de barras
fig, ax = plt.subplots()

# Gráfica de producción
ax.bar(meses - 0.3, produccion, width=0.3, label='Producción', color='blue')

# Gráfica de inventarios
ax.bar(meses, inventario, width=0.3, label='Inventario', color='orange')

# Gráfica de demanda
ax.bar(meses + 0.3, demanda, width=0.3, label='Demanda', color='green')

# Etiquetas y título
ax.set_xlabel('Meses')
ax.set_ylabel('Cantidad de ventanas')
ax.set_title('Producción óptima, inventario óptimo y demanda por mes')
ax.set_xticks(meses)
ax.legend()

# Mostrar el gráfico
plt.show()
```
